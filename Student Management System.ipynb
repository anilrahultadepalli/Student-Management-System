```python
import mysql.connector
from tkinter import *
from tkinter import messagebox

def create_connection():
    try:
        connection = mysql.connector.connect(host="localhost", user="root", password="21KN1A05G9", database="rahul")
        return connection
    except mysql.connector.Error as e:
        messagebox.showerror("Error", f"Error connecting to MySQL: {str(e)}")
        return None

def adds(name, rollno, grade):
    connection = create_connection()
    if connection:
        cur = connection.cursor()
        s = "INSERT INTO studpy(name, rollno, grade) VALUES(%s,%s,%s);"
        values = (name, rollno, grade)
        cur.execute(s, values)
        connection.commit()
        connection.close()
        messagebox.showinfo("Success", "Student added successfully!")

def updates(name, rollno, grade):
    connection = create_connection()
    if connection:
        cur = connection.cursor()
        s = "UPDATE studpy SET name=%s, rollno=%s, grade=%s WHERE rollno=%s;"
        values = (name, rollno, grade, rollno)
        cur.execute(s, values)
        connection.commit()
        connection.close()
        messagebox.showinfo("Success", "Student updated successfully!")

def deletes(rollno):
    connection = create_connection()
    if connection:
        cur = connection.cursor()
        s = "DELETE FROM studpy WHERE rollno=%s;"
        values = (rollno,)
        cur.execute(s, values)
        connection.commit()
        connection.close()
        messagebox.showinfo("Success", "Student deleted successfully!")

def show():
    connection = create_connection()
    if connection:
        cur = connection.cursor()
        cur.execute("SELECT * FROM studpy;")
        rows = cur.fetchall()
        for row in rows:
            listbox.insert(END, row)
        connection.close()

def add_student():
    adds(entry_name.get(), entry_rollno.get(), entry_grade.get())
    entry_name.delete(0, END)
    entry_rollno.delete(0, END)
    entry_grade.delete(0, END)

def update_student():
    updates(entry_name.get(), entry_rollno.get(), entry_grade.get())
    entry_name.delete(0, END)
    entry_rollno.delete(0, END)
    entry_grade.delete(0, END)

def delete_student():
    deletes(entry_rollno.get())
    entry_name.delete(0, END)
    entry_rollno.delete(0, END)
    entry_grade.delete(0, END)

def view_students():
    listbox.delete(0, END)
    show()

# GUI setup
root = Tk()
root.title("Student Management System")

Label(root, text="Name").grid(row=0, column=0)
entry_name = Entry(root)
entry_name.grid(row=0, column=1)

Label(root, text="Roll No").grid(row=1, column=0)
entry_rollno = Entry(root)
entry_rollno.grid(row=1, column=1)

Label(root, text="Grade").grid(row=2, column=0)
entry_grade = Entry(root)
entry_grade.grid(row=2, column=1)

Button(root, text="Add Student", command=add_student).grid(row=3, column=0, pady=5)
Button(root, text="Update Student", command=update_student).grid(row=3, column=1)
Button(root, text="Delete Student", command=delete_student).grid(row=4, column=0, pady=5)
Button(root, text="View Students", command=view_students).grid(row=4, column=1)

listbox = Listbox(root, width=50)
listbox.grid(row=5, column=0, columnspan=2)

root.mainloop()
```
